```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ex1 Scatter Chart

#set the number of points
n1=1024 

# creat x and y by random
x1=np.random.normal(0,1,n1)  
y1=np.random.normal(0,1,n1)

# for color value
T1=np.arctan2(x1,y1)

#  s: size; c: color; alpha is 50%
plt.scatter(x1,y1,s=50,c=T1,alpha=0.5)

#  set the range of figure
plt.xlim(-2,2)
plt.ylim(-2,2)

# ignore x and y ticks
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Ex2 Bar chart

n2=12
x2=np.arange(n2)
y21=(1-x2/float(n2))*np.random.uniform(0.5,1.0,n2)  #random creat n real number between0.5 ti 1.0
y22=(1-x2/float(n2))*np.random.uniform(0.5,1.0,n2)

plt.bar(x2,+y21, facecolor='red',edgecolor='black')
plt.bar(x2,-y22, facecolor='black', edgecolor='red')

plt.xlim(-3,n2)
plt.ylim(-1.25,1.25)

for x,y in zip(x2,y21):
    plt.text(x,y+0.05,'%.1f' %y,ha='center',va='bottom')
#     ha: horizontal; va: vertical
for x,y in zip(x2,y22):
    plt.text(x,-y-0.05,'%.1f' %y,ha='center',va='top')
    
plt.show()
```
