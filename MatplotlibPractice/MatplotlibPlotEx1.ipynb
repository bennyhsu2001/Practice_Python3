```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 =pd.Series(np.random.randn(1000),index=np.arange(1000))
data1=data1.cumsum()
data1.plot()
plt.show()
```

```python
data2=pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list('ABCD'))
print(data2.head(5))
data2.plot()
plt.show()
```

```python
data2=data2.cumsum()
print(data2.head(5))
data2.plot()
plt.show()
```

```python
class1=data2.plot.scatter(x='A',y='B',color='Blue',label='Class1')
data2.plot.scatter(x='A',y='C',color='Green',label='Class2',ax=class1)

plt.show()
```
