```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,50)
#  creat a line with 50 points from -3 to 3 
```

```python
y1 = 2*x+1

plt.figure()
plt.plot(x,y1)
plt.show()
```

```python
y2 = x**2

plt.figure()
plt.plot(x,y2)
plt.show()
```

```python
plt.figure(num=10,figsize=(8,6))
    # figsize used to set the size of figure you wanna show
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=5.0,linestyle='--',)
plt.show()
```

```python
plt.figure()
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')
plt.plot(x,y2)

#  use xlim and ylim to set the part of figure you wanna show
plt.xlim((-1,2))
plt.ylim((-2,3))

#  to name the axis
plt.xlabel('I am x')
plt.ylabel('I am y')

plt.show()
```

```python
plt.figure()
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')
plt.plot(x,y2)
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('I am x')
plt.ylabel('I am y')
# -------------------------------------------

new_ticks=np.linspace(-1,2,5)
print(new_ticks)

#  set the label that u need
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,1,2.7,3],['suck','worse','not bad','good','wonderful'])

plt.show()
```

```python
plt.figure()
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')
plt.plot(x,y2)
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('I am x')
plt.ylabel('I am y')
# --------------------------------------------

ax=plt.gca()
ax.spines['right'].set_color('green')
ax.spines['left'].set_color('orange')
ax.spines['top'].set_color('none')
#  the button side is still default value
plt.show()
```

```python
plt.figure()
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--',label='linear line')
plt.plot(x,y2,label='square line')
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.legend(loc='lower right')

plt.show()
```

```python
plt.figure()
# setting label 'l1,' must have ',' because plt.plot return a list
l1, = plt.plot(x, y1, label = 'line')
l2, = plt.plot(x, y2, label = 'parabola', color = 'red', linewidth = 1.0, linestyle = '--')
# set the scale of figure
plt.xlim((-1, 2))
plt.ylim((-2, 3))
# set the axis' lable
plt.xlabel('i am x')
plt.ylabel('i am y')
# 设置x坐标轴刻度, 原来为0.25, 修改后为0.5
plt.xticks(np.linspace(-1, 1, 5))
# 设置y坐标轴刻度及标签, $$是设置字体
plt.yticks([0, 0.5], ['$minimum$', 'normal'])
# set the legend
plt.legend(handles = [l1, l2,], labels = ['a', 'b'], loc = 'best')
plt.show()
```
