```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x=np.linspace(-1,1,50)
y=2*x+1
```

```python
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y)

ax=plt.gca()
#  clear right and top edges
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#  to move x and y axis
#  set the origin x equal 0
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#  set the origin y equal 0
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


# set the dash line: ax's plot x is x0; y is from 0 to y0 
x0=1
y0=2*x0+1
ax.plot([x0,x0],[0,y0],'--',linewidth=2.5)

#  set the spot: s is the size of spot
plt.scatter([x0,x0],[y0,y0],s=50,color='red')

# arrowprops is the setting of arrow; arrowstyle is the direction
plt.annotate('2x+1=%s' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# set another annatation
plt.text(-1.0,1.0,'this is the test text', fontdict={"color":"red","size":16})

plt.show()
```

```python
#  another example

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

theta = np.arange(0, 2*np.pi, 2*np.pi/100)
ax.plot(np.cos(theta), np.sin(theta))

plt.style.use('ggplot')
ax.set_xticks([-1.2, 1.2])
ax.set_yticks([-1.2, 1.2])

plt.show()
```
