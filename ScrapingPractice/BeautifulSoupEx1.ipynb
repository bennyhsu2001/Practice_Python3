```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
html1=urlopen('https://morvanzhou.github.io/static/scraping/basic-structure.html').read().decode('utf-8')
soup1 = BeautifulSoup(html1,features='lxml')
print(soup1)
```

```python
html2=urlopen('https://www.iii.org.tw/').read().decode('utf-8')
soup2 = BeautifulSoup(html2,features='lxml')
print(soup2)
```

```python
print(soup2.p)
```

```python
all_href2=soup2.find_all('a')
for item in all_href2:
    print(item)
```

```python
for item in all_href2:
    print(item['href'])
```

```python
all_href1 = soup1.find_all('a')
all_href1 = [l['href'] for l in all_href1]
print('\n', all_href1)
```

```python
html3 = urlopen("https://morvanzhou.github.io/static/scraping/list.html").read().decode('utf-8')
print(html3)
```

```python
soup3=BeautifulSoup(html3,features='lxml')
month = soup3.find_all('li',{'class':'month'})
print(month,'\n')

for m in month:
    print(type(m))
    print(m)
    print(m.get_text(),'\n')
```
