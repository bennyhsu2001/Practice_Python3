```python
from urllib.request import urlopen
import re
```

```python
html1=urlopen('https://morvanzhou.github.io/static/scraping/basic-structure.html').read().decode('utf-8')
print(html1)
```

```python
html2=urlopen('https://www.iii.org.tw/').read().decode('utf-8')
print(html2)
```

```python
res1=re.findall("<title>(.+?)</title>",html1)
print(res1)
```

```python
res2=re.findall(r"<p>(.*?)</p>",html2,flags=re.DOTALL)
for item in res2:
    print(item)
```

```python
res2=re.findall('href="(.*?)"',html2)
for item in res2:
    print(item)
```
