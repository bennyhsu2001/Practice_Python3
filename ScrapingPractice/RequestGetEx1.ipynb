```python
import requests
```

Ex1  to get original html code

```python
url = 'http://www.e-happy.com.tw'
html1 = requests.get(url)
html1.encoding='utf-8'
if html1.status_code == requests.codes.ok:
    # requests.codes.ok means the connect and request is ok
    print(html1.text)
```

Ex2  add url

```python
payload={'key1':'value1','key2':'value2'}
html2 = requests.get('http://httpbin.org/get',data=payload)
print(html2.url)
```
