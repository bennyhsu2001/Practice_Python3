```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine

def get_html(url):
    info = []
    try:
        r = requests.get(url,timeout = 30)
        soup = BeautifulSoup(r.text,'html.parser')
#         print(soup)    # get the html text
#         print(soup.select('.book-mid-info a'))    # it's a list contain all tags
        for i in soup.select('.book-mid-info a'):
        #use ccs select method, return all 'a' tags which xalss are 'book-mid-info'
            info.append(i.text)  
#         print(info)
        return info
    except:
        print('a error')

def transform_df(list):  
    
    n = len(list)#每页有i个排名，info共4i个元素
    title =list[0:n:4]
    author =list[1:n:4]
    style =list[2:n:4]
    updata = list[3:n:4]
    #直接转换成DataFrame数表是以行为标准写入（4xi），需要转置成（ix4）数表
    df = pd.DataFrame([title,author,style,updata]).T          
    return df

def main():
    info = []
    for i in range(1,11):
        url = 'http://www.qidian.com/rank/hotsales?page='+str(i)
        info += get_html(url)
    print(info)
    
    data = transform_df(info)
    
    data.columns =['书名','作者','风格','最新章节']
    #data.to_csv('D:\Paperwork\PythonData\qidian_rank.csv')
    #print(data)
    
    engine = create_engine('sqlite:///testDB')
    data.to_sql('qidian_rank',con = engine,if_exists ='replace',index = False)
    print('Done')
    
main() 
```
