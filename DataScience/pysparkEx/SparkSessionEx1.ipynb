```python
import pandas as pd
import numpy as np
import time
```

```python
from pyspark.sql import SparkSession

# local mode
spark = SparkSession\
        .builder\
        .appName("mldemo")\
        .getOrCreate()
```

```python
# Check SparkSession is OK
spark.sparkContext.appName
```

```python
s = time.time()
df = spark.read.csv("../data/performance_detail.csv",sep = ';', header = True, inferSchema = True)
print("read csv cost time:", time.time()-s)
```
