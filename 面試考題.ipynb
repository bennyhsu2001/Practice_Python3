# Correlation

```python
import math

def avg_list(A):
    s =[0 for i in range(len(A[0]))]
    for row in A:
        for i in range(len(row)):
            s[i] += row[i]
    return [i/float(len(A)) for i in s]


def corr_cal(A, i, j, avg_list):
    x = y = s = 0.0
    for row in A:
        x += (row[i]-avg_list[i])**2
        y += (row[j]-avg_list[j])**2
        s += (row[i]-avg_list[i])*(row[j]-avg_list[j])
#     print(x, y, s)
    r = s/((math.sqrt(x))*(math.sqrt(y)))
    return r


def corr_matrix(A):
    corr = [[0 for i in range(len(A[0]))] for i in range(len(A[0]))]
    cnt = 0
    avg = avg_list(A)
    for row in corr:
        cnt2 = 0
        for item in row:
            corr[cnt][cnt2] = corr_cal(A, cnt, cnt2, avg)
            cnt2 += 1
        cnt += 1
    return corr

# ----------------------------------------------------------

A = [['3','4','6'],
     ['12','8','3'],
     ['6','4','9'],
     ['0','11','15'],
     ['8','7','7']]

A = [[int(i.encode('utf-8')) for i in A[j]] for j in range(len(A))]

# avg_list = avg_list(A)     
# avg_list

corr = corr_matrix(A)
corr
```

# No three consecutive letters


```python
def solution(A, B):
    if A>=B: 
        a,b = A,B;
        x,y = 'a','b'
    if A<B: 
        a,b = B,A;
        x,y = 'b','a'
    
    if a >= (2*b+3):
        return 'out of range'
    elif  a>=b and b>0 and a//b<3:
        ans = (x*(a//b) + y)*b + x*(a-b*(a//b))
        return ans
    elif a//b>=3 and b<=2:
        ans = ((x+x+y)*b) + (x*(a-2*b))
        return ans
    else:
        return 0
```

# Tickets Needed To get Minimum cost




```python
def solution(A):
    from operator import itemgetter

    d=[]
    for i in range(30):
        d.append(1 if (i+1) in A else 0)

    cost = 0
    while True:
        l = len(d)
        result = []
        for i in range(l):
            s = sum(d[i:i+7])
            result.append((i,s))          
        
        
        result.sort(key = itemgetter(1))
        b = result[-1][1]
        if b < 4:
            break
        for j in result:
            if j[1] == b:
                k = j
                break
        d[k[0]:k[0]+7] = [0]*7
        cost = cost + 7 
    cost = cost + d.count(1) * 2
    ans = min(25,cost)
    return ans
```
